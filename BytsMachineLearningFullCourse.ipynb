```python
import numpy as np
import pandas as pd
#data preparation
#handling missing values
#normalization,standardization,outliers
df=pd.read_csv("https://raw.githubusercontent.com/nunnarilabs/ml/master/Automobile_Data/automobile_data.csv")
df.head()
```

```python
df.info()
```

```python
df.replace("?",np.nan,inplace=True)
df.head()
```

```python
df["normalized-losses"].unique()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
avg=df["normalized-losses"].astype("float").mean(axis=0)
print(avg)
```

```python
df["normalized-losses"].replace(np.nan,avg,inplace=True)
df
```

```python
meanOfBore=df["bore"].astype(float).mean(axis=0)
print(meanOfBore)
```

```python

df["bore"].replace(np.nan,meanOfBore,inplace=True)
df
```

```python
df["num-of-doors"].value_counts()
```

```python
k=df["num-of-doors"].mode()
print(k)


```

```python
df["num-of-doors"].replace(np.nan,'k',inplace=True)
df
```

```python
df.dropna(subset=["price"],axis=0,inplace=True)
df.reset_index(drop=True,inplace=True)
df
```

```python
df["stroke"].replace(np.nan,df['stroke'].astype('float').mean(axis=0),inplace=True)
df["horsepower"].replace(np.nan,df['horsepower'].astype('float').mean(axis=0),inplace=True)
df["peak-rpm"].replace(np.nan,df['peak-rpm'].astype('float').mean(axis=0),inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df[['bore','stroke']]=df[['bore','stroke']].astype('float')
df[['normalized-losses']]=df[['normalized-losses']].astype('int')
df[['price']]=df[['price']].astype('float')
df[['peak-rpm']]=df[['peak-rpm']].astype('float')
df[['horsepower']]=df[['horsepower']].astype('float')
df
```

```python
from scipy import stats
hp_zscore=np.abs(stats.zscore(df['normalized-losses']))#z=x-miu/standard deviation
hp_zscore
```

```python
thre=hp_zscore.max()
index_val=np.where(hp_zscore>=thre)
df.iloc[index_val]['normalized-losses']
```

```python
nd=df.drop(index=df.iloc[index_val]['normalized-losses'].index)
nd
```
